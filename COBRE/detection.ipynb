### Applied to line

```python
from scipy import signal
sig = np.repeat([0., 1., 1., 0., 1., 0., 0., 1.], 128)
sig_noise = sig + np.random.randn(len(sig))
corr = signal.correlate(sig_noise, np.ones(128), mode='same') / 128
```

```python
import matplotlib.pyplot as plt
clock = np.arange(64, len(sig), 128)
fig, (ax_orig, ax_noise, ax_corr) = plt.subplots(3, 1, sharex=True)
ax_orig.plot(sig)
ax_orig.plot(clock, sig[clock], 'ro')
ax_orig.set_title('Original signal')
ax_noise.plot(sig_noise)
ax_noise.set_title('Signal with noise')
ax_corr.plot(corr)
ax_corr.plot(clock, corr[clock], 'ro')
ax_corr.axhline(0.5, ls=':')
ax_corr.set_title('Cross-correlated with rectangular pulse')
ax_orig.margins(0, 0.1)
fig.tight_layout()
fig.show()
```

```python
from PIL import Image
from pylab import *
import numpy as np



def LoadProcessImg(inputImg):
    pilImage=Image.open(inputImage)
    #pilImage = pilImage.convert('1') 
    #pilImage.draft('L',(500,500))
    img= np.asarray(pilImage)


    #imshow(pilImage)

    plt.figure()
    plt.subplot(111,aspect='equal')
    plt.pcolormesh(np.flipud(img[:,:,0]))
    
    new = np.zeros([img.shape[0],img.shape[1]])
    new+=0.2989*img[:,:,0]
    new+=0.5870*img[:,:,1]
    new+=0.1140*img[:,:,2]
    myo = new

    # Normalize image
    myo -= np.min(myo)
    myo /= np.max(myo)
    
    # invert
    #myo = 1-myo
    
    plt.subplot(111,aspect='equal')


    plt.pcolormesh(np.flipud(myo),cmap="gray")    
    return myo

```

```python
# from smooth.py
def shift2d(a,shift):
  ash=np.roll(np.roll(a,shift,axis=0),shift,axis=1)

  return ash
```

```python


from scipy import fftpack as fftp

def DoCorr(m, s,label=""):
    ###############
    sPadded = np.zeros( m.shape )
    dim = np.shape(s)[0]
    sPadded[0:dim,0:dim] = s
    sPaddedS = shift2d(sPadded,-dim/2)

    #plt.figure()
    #plt.subplot(111,aspect='equal')
    #plt.title("s")
    #pcolormesh(sPadded,cmap='gray')
    #pcolormesh(sPadded,cmap='gray')
    #xlim([0,2*dim])
    #ylim([0,2*dim])
    

    ########
    ms=fftp.fftshift(m)
    M = fftp.fft2(ms)
    S = fftp.fft2(sPaddedS)
    MS = M*S
    mss = fftp.ifft2(MS)
    ms = fftp.ifftshift(mss)
    ms = np.real(ms)
    
    plt.figure()
    plt.subplot(121,aspect='equal')
    plt.title("Input [%s]"%label)
    pcolormesh(m,cmap='gray')
    plt.axis('off')
    plt.subplot(122,aspect='equal')
    plt.title("Result")
    pcolormesh(ms,cmap='gray')
    plt.axis('off')
    return ms
   
```

```python
def evalCorr(response):
    evalLine = response[dim/2,:]
    #plt.plot(evalLine)
    args = np.argwhere( evalLine > 0 )
    return  np.sum( evalLine[args])
```

```python
# Create Image and Filter
###############
#m = np.zeros([256,768])
#m[0:10,0:10] = 1
dim = 64
s = sham[0:dim,0:dim]

truth=DoCorr(sham,s,label="SHAM") 
truth -= np.mean(truth)
# max/std dev not the best, since underlying surface is periodic
#print np.max(truth), np.std(truth), np.log(np.max(truth)/np.std(truth))
print evalCorr(truth)


incorrect=DoCorr(chf,s,label="CHF") 
incorrect -= np.mean(incorrect)
#print np.max(x), np.std(x), np.log(np.max(x)/np.std(x))
print evalCorr(incorrect)


null=DoCorr(rando,s, label="Random") 
null -= np.mean(null)
#print np.max(x), np.std(x), np.log(np.max(x)/np.std(x))
print evalCorr(null)




plt.figure()
plt.plot(truth[dim/2,:],label="Truth")
plt.plot(incorrect[dim/2,:],label="CHF")
plt.plot(null[dim/2,:],label="random")
plt.legend()
#plt.plot(np.sum(truth,axis=0),label="Truth")
#plt.plot(np.sum(x,axis=0),label="CHF")
```
