```python
zeta = 1. # persistence length [nm]
Ang_to_nm = 1e-1
D = 20. *Ang_to_nm   # binding site distance [nm]
L = 12. * Ang_to_nm  # tether length [nm]

def G_DZL(zeta,D,L):
    prefac = sqrt(3/(4*np.pi*zeta*L)) 
    prefac =prefac*prefac*prefac
    expon = -3*D*D/(4*zeta*L)
    return prefac * exp(expon)
```
